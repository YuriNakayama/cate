```python
from sklift.datasets.datasets import (
    fetch_megafon
)
import pandas as pd
import numpy as np
```

```python
ds = fetch_megafon()
```

```python
df = pd.merge(ds.data, ds.target, left_index=True, right_index=True)
df = pd.merge(df, ds.treatment, left_index=True, right_index=True)
```

```python
df
```

```python
def onehot_encoding(df: pd.DataFrame, columns: list[str]) -> pd.DataFrame:
    _df = df.copy()
    for column in columns:
        _df = pd.merge(
            _df.drop(column, axis=1),
            pd.get_dummies(df[column], prefix=column, dtype=int),
            left_index=True,
            right_index=True,
        )
    return _df
```

```python
def label_encoding(df: pd.DataFrame, columns: list[str]) -> pd.DataFrame:
    _df = df.copy()
    for column in columns:
        _df[column] = _df[column].astype("category").cat.codes
    return _df
```

```python
label_encoding(df, ["history_segment", "zip_code", "channel"])
```
